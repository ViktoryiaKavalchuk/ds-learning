```python
# найти коэфициенты, без линейной регрессии

quantile_step_percent = 5
row_headers = [f"{step}%" for step in range(quantile_step_percent, 101, quantile_step_percent)]

quantile_list = []
for column_name in column_names:
    row = []
    for quantile in numpy.linspace(quantile_step_percent / 100, 1, len(row_headers)):
        value = float(numpy.quantile(data[column_name], quantile))
        row.append(value)
    quantile_list.append(row)

quantile_table = pandas.DataFrame(quantile_list, index=column_names, columns=row_headers).T.map(lambda x: f"{x:.6f}")

quantile_table
```

CRIM - per capita crime rate by town
    - moderate negative (-0.39) influence (70% has low value)

ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
    - moderate positive (0.36) influence (75% has 0 value, the other have values starting from 12 till 80)
    
INDUS - proportion of non-retail business acres per town.
    - strong negative (-0.48) influence (but 30% has no big change in value)

CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
    - light positive (0.18) influence (95% has 0 value)
    
NOX - nitric oxides concentration (parts per 10 million)
    - strong negative (-0.43) influence

RM - average number of rooms per dwelling
    - very strong positive (0.70) influence

AGE - proportion of owner-occupied units built prior to 1940
    - moderate negative (-0.38) influence

DIS - weighted distances to five Boston employment centres
    - less moderate positive (0.25) influence

RAD - index of accessibility to radial highways
    - moderate negative (-0.38) influence

TAX - full-value property-tax rate per $10,000
    - strong negative (-0.47) influence

PTRATIO - pupil-teacher ratio by town
    - strong negative (-0.51) influence

B - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town
    - moderate positive (0.33) influence

LSTAT - % lower status of the population
    - very strong negative (-0.74) influence

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_dict = {}

def build_parameter_dependency_graph(x_column: str):
    sns.regplot(x=x_column, y=data['MEDV'], data=data, scatter_kws={'alpha':0.6})
    
    correlation = data[x_column].corr(data['MEDV'])
    print(f"{x_column} -> MEDV correlation: {correlation:.2f}")

    correlation_dict.setdefault(x_column, round(float(correlation), 2))
```

```python
#AI proposal


lstat = numpy.array(data['LSTAT'])
rm = numpy.array(data['RM'])
ptratio = numpy.array(data['PTRATIO'])
indus = numpy.array(data['INDUS'])
tax = numpy.array(data['TAX'])
nox = numpy.array(data['NOX'])
crim = numpy.array(data['CRIM'])
age = numpy.array(data['AGE'])
rad = numpy.array(data['RAD'])
zn = numpy.array(data['ZN'])
b = numpy.array(data['B'])
dis = numpy.array(data['DIS'])
chas = numpy.array(data['CHAS'])

medv = numpy.array(data['MEDV'])

bias_ratio = numpy.ones_like(medv)
feature_matrix = numpy.column_stack(
    (
        lstat,
        rm,
        ptratio,
        indus,
        tax,
        nox,
        crim,
        age,
        rad,
        zn,
        b,
        dis,
        chas,
        bias_ratio
    )
)
theta = numpy.linalg.inv(feature_matrix.T @ feature_matrix) @ feature_matrix.T @ medv

lstat_ratio, rm_ratio, ptratio_ratio, indus_ratio, tax_ratio, nox_ratio, crim_ratio, age_ratio, rad_ratio, zn_ratio, b_ratio, dis_ratio, chas_ratio, bias_ratio_ratio = theta

print(
    f"lstat_ratio =\t\t{lstat_ratio},\n"
    f"rm_ratio =\t\t{rm_ratio},\n"
    f"ptratio_ratio =\t\t{ptratio_ratio},\n"
    f"indus_ratio =\t\t{indus_ratio},\n"
    f"tax_ratio =\t\t{tax_ratio},\n"
    f"nox_ratio =\t\t{nox_ratio},\n"
    f"crim_ratio =\t\t{crim_ratio},\n"
    f"age_ratio =\t\t{age_ratio},\n"
    f"rad_ratio =\t\t{rad_ratio},\n"
    f"zn_ratio =\t\t{zn_ratio},\n"
    f"b_ratio =\t\t{b_ratio},\n"
    f"dis_ratio =\t\t{dis_ratio},\n"
    f"chas_ratio =\t\t{chas_ratio},\n"
    f"bias_ratio_ratio =\t{bias_ratio_ratio}"
)
```

```python
def calculate_house_price(lstat: float, rm: float, ptratio: float, indus: float, tax: float, nox: float, crim: float, age: float, rad: float, zn: float, b: float, dis: float, chas: float) -> float:
    return float(
        lstat_ratio * lstat 
        + rm_ratio * rm 
        + ptratio_ratio * ptratio
        + indus_ratio * indus
        + tax_ratio * tax
        + nox_ratio * nox
        + crim_ratio * crim
        + age_ratio * age
        + rad_ratio * rad
        + zn_ratio * zn
        + b_ratio * b
        + dis_ratio * dis
        + chas_ratio * chas
        + bias_ratio_ratio
    )

data['MEDV_PREDICTED'] = data.apply(
    lambda row: calculate_house_price(
        row['LSTAT'], row['RM'], row['PTRATIO'], row['INDUS'], row['TAX'],
        row['NOX'], row['CRIM'], row['AGE'], row['RAD'], row['ZN'],
        row['B'], row['DIS'], row['CHAS']
    ),
    axis=1
)

data['MEDV - MEDV_PREDICTED'] = (data['MEDV'] - data['MEDV_PREDICTED']).round(2).abs()
data['MEDV - MEDV_PREDICTED SQUARE'] = (data['MEDV'] - data['MEDV_PREDICTED']).round(2) ** 2
data.describe()
```
